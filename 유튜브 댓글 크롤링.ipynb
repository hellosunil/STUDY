```python
search = input('유튜브에서 검색할 단어를 입력하세요 : ')
keyword = urllib.parse.quote(search)
url = f'https://www.youtube.com/results?search_query={keyword}&sp=EgIQAQ%253D%253D'
driver = webdriver.Chrome('chromedriver.exe')
driver.implicitly_wait(1)
driver.get(url)
driver.maximize_window()
time.sleep(3)
html = driver.page_source
soup = BeautifulSoup(html, 'lxml')
body = driver.find_element_by_css_selector('body')
for i in range(2):
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(1)
    
tumbnail = driver.find_elements_by_tag_name('ytd-video-renderer')

result = []

for i in tumbnail:
    try:
        time.sleep(1)
        i.click()
        time.sleep(1)
        tbody = driver.find_element_by_tag_name('body')
        result_list = []

        for j in range(2):
            tbody.send_keys(Keys.PAGE_DOWN)
            time.sleep(2)

        html = driver.page_source
        soup = BeautifulSoup(html, 'lxml')
        reply = soup.find_all('yt-formatted-string', {'id' : 'content-text'})
        reply_result = ''
        for k in reply:
            reply_result = reply_result + k.text + ' '

        title = soup.find('h1', {'class':'title style-scope ytd-video-primary-info-renderer'})
        ttitle = title.text

        result_list.append(ttitle)
        result_list.append(reply_result)

        result.append(result_list)
    except:
        print('오류')

    driver.back()

result1 = pd.DataFrame(result)
result1.to_csv('data/f"{search}"유튜브댓글.csv')
```
