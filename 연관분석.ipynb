```python
# 1인 구매 리스트
dataset = [
    ['커피', '녹차', '얼음'],
    ['커피', '얼음', '우유'],
    ['커피', '딸기'],
    ['바나나', '녹차', '딸기'],
    ['커피', '얼음'],
    ['우유','커피'],
    ['우유', '딸기'],
    ['커피'],
    ['얼음', '커피']
]

import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
te_ary
```

### 전체 물품 별 구매사항을 bool로 정리

```python
df = pd.DataFrame(te_ary, columns = te.columns_)
df
```

```python
# 지지도(support) : 전체거래항목 중 a,b를 포함하는 비율
# 신뢰도(confidence) : a를 포함하는 거래 중에 a와 b가 동시에 거래되는 비중
# 향상도(lift) : a와 b가 독립 = 1, 1보다 크면 연관성이 높다, 1보다 작으면 연관성이 낮다

f_data = apriori(df, min_support = 0.2, use_colnames=True)#지지도가 0.2이상인 것만 보여줘
f_data
```

```python
from mlxtend.frequent_patterns import association_rules

association_rules(f_data, metric='confidence', min_threshold=0.3)
```
