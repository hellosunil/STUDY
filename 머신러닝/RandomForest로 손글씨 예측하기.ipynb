#### 데이터베이스에서 0 ~ 9 까지의 손글씨 데이터(col_0 ~ col_63로 디지털화, col_64는 손글씨의 정답)
#### 필요 없는 1, 7을 제외한 데이터를 가져와서 ndarray로 만들기

```python
# 필요 모듈 import

import pandas as pd
import numpy as np
import pymysql
from sqlalchemy import text, create_engine
```

```python
# 데이터베이스 접속
protocol = 'mysql+pymysql'
user = 'bigdata'
passwd = 'bigdata1234'
host = 'localhost'
port = '3306'
db = 'bigdata'
conn = create_engine(f'{protocol}://{user}:{passwd}@{host}:{port}/{db}')
```

```python
sql = '''
SELECT *
FROM tb_digit
WHERE col_target NOT IN (1, 7)
'''
df = pd.read_sql(sql, conn)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
X = df.iloc[:,:-1]
Y = df.iloc[:,-1].values.reshape(-1, 1)
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2)
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(X_train, Y_train)
accuracy_score(rf_model.predict(X_test), Y_test)
```
