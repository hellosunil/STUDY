##### 군집

##### K-means Clustering

```python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris['data']
y = iris['target']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

from sklearn.cluster import KMeans
k = 3
kmeans = KMeans(n_clusters = k)
y_pred = kmeans.fit_predict(X)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y, y_pred)
```

```python
sum(y==y_pred)/len(y)
```

##### KMeans 속도 개선과 미니배치 KMeans
- 삼각 부등식을 사용해 불필요한 거리계산을 줄임(AC <= AB + BC) : 두 점 사이의 거리를 항상 가장 짧은 거리가 되게 설정
- 이는 KMeans 클래스에서 기본적으로 사용하는 방식임

- 각 반복마다 미니배치를 사용해 센트로이드를 조금씩 이동하는 방식
- 일반적으로 알고리즘 속도를 3배에서 4배정도 높임
- 또한 대량의 데이터셋에 군집 알고리즘을 적용할 수도 있게 됨

```python
from sklearn.cluster import MiniBatchKMeans
minibatch_kmeans = MiniBatchKMeans(3)
#minibatch_kmeans.fit(X)
y_pred = minibatch_kmeans.fit_predict(X)
```

```python
# 성능은 더 떨어짐
accuracy_score(y, y_pred)
```

- -1 ~ +1까지의 값을 가짐

```python
from sklearn.metrics import silhouette_score
silhouette_score(X, kmeans.labels_)
```

```python
for i in range(2, 10):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(X)
    print(i," : ", silhouette_score(X, kmeans.labels_))
```

##### DBSCAN
- 밀도 기반 클러스터링
- predict 매서드를 제공하지 않음
- 다시 말해, 새로운 샘플에 대해 클러스터를 예측할 수 없음
- 예측기를 따로 만들어줘야 함

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=1000, noise=.05)
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan.fit(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=dbscan.labels_)
```

```python
# 예측기 생성
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(50)
knn.fit(dbscan.components_, dbscan.labels_[dbscan.core_sample_indices_])
```

```python
import numpy as np
X_new = np.array([[-0.5,0],[0,0.5],[1,-0.1],[2,1]])
knn.predict(X_new)
```
