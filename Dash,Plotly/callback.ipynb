```python
# dash에 필요한 라이브러리 
import numpy as np
import pandas as pd

import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output


Myapp = dash.Dash(__name__)

Myapp.layout = html.Div(
    [
        html.H1("우리의 처음 대시"),
        html.Hr(),
        dcc.Checklist(id="aaa1",
                    options = [
                        {"label":1,"value":1},
                        {"label":2,"value":2},
                    ]),
        dcc.RadioItems(id="aaa2",
                      options = [
                          {"label":"apple","value":"사과"},
                          {"label":"graph","value":"포도"},
                      ]),
        html.Div(id="output-aaa2"),
        html.Div(id="output-aaa3"),
        dcc.Dropdown(id="aaa3",
                    options = [
                        {"label":"사과","value":"apple"},
                        {"label":"포도","value":"graph"},
                    ]),
    ]  
)

@Myapp.callback(
    Output("output-aaa2","children"), # Response
    Input("aaa2","value"), # Request
)
def radio_update(fruit):
    return html.H1(fruit)

@Myapp.callback(
    Output("output-aaa3","children"),
    Input("aaa3","value")
)
def drop_update(fruit):
    return html.H1(fruit)


Myapp.run_server(host="localhost",port=7777)
```
