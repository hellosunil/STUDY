```python
# dash에 필요한 라이브러리
# callback = f와 b의 수신
# 정적인 화면 구성 시, callback이 필요 없음
import numpy as np
import pandas as pd

import dash
import dash_core_components as dcc # f -> b으로 보낼 때
import dash_html_components as html # html 화면 구성 f-b 수신 없는 경우
from dash.dependencies import State, Input, Output
```

```python
app = dash.Dash(__name__)

# layout 그려주기
app.layout = html.Div(
    [
        html.H1('우리의 처음 Dash'),
        html.Hr(),
        dcc.Checklist(id='aaa1',  # unique한 id값 설정
                    options = [
                        {'label':1, 'value':1},
                        {'label':2, 'value':2}]),
        dcc.RadioItems(id='aaa2',
                      options = [
                          {'label':'apple', 'value':'사과'},
                          {'label':'orange', 'value':'오렌지'}]),
        ## radioitems를 받아서
        html.Div(id='output-aaa2'),
        
        dcc.Dropdown(id='aaa3',
                       options = [
                          {'label':'사과', 'value':'apple'},
                          {'label':'오렌지', 'value':'orange'}]),
        html.Div(id='output-aaa3')
    ]
)



@app.callback(
    Output('output-aaa2','children'), ## 어디에 #Response
    Input('aaa2','value')  ## 뭘 받아서 Request
)

## 어떻게 처리할까
def radio_update(fruit):
    return html.H1(fruit) #children으로 보내고 children은 output-aaa2로 보냄


@app.callback(
    Output('output-aaa3', 'children'),
    Input('aaa3', 'value')
)

def radio_update(fruit):
    return html.H1(fruit) #children으로 보내고 children은 output-aaa2로 보냄













app.run_server(host='localhost', port=6677)
```
