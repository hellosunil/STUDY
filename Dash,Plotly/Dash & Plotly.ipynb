```python
# 화면꾸미기 (frontend)
app = dash.Dash(external_stylesheets=[dbc.themes.SOLAR]) # app instance 
app.layout = html.Div(
    [
        html.H1("IRIS-Clustering"),
        html.Hr(),
        dcc.Dropdown(id="cluster-number",
                 options = [{"label":str(i)+"클러스터수","value":i}  for i in np.arange(2,12)]
                    ),
        dcc.Dropdown(id="input-columns",
                     options =[{"label":i,"value":i} for i in df.columns[:-1]]
                    ),
        html.Hr(),
        html.Div(id="gr-mean"),
    ]

)
# 화면에다 output보내기(backend)
@app.callback(
    Output("gr-mean","children"),
    Input("cluster-number","value"),
    Input("input-columns","value"),
)
def update_html(n,col):
    km = KMeans(n)
    km.fit(df.iloc[:,:-1])
    gr = km.labels_
    dt = df.copy()
    dt['gr'] = gr
    x = dt.groupby(["gr"]).sum().index
    y = dt.groupby(["gr"]).sum().loc[:,col]
    
    x = x.astype("str")
    colors = ['lightslategray',] * n
    colors[np.argmax(y.values)] = 'crimson'
    
    title = "Iris 컬럼중 "+col+" 컬럼의 "+str(n)+" 개의 군집들의 평균"
    fig = go.Figure(data=[go.Bar(
    x = x,
    y = y,
    marker_color=colors
    )])
    fig.update_layout(title_text=title)
    
    group = x[np.argmax(y.values)]
    m = y[np.argmax(y.values)]
    html_str = []
    for i in pd.Series(y.index).astype("str") + " : "+ y.astype("str"):
        html_str.append(i)
    html_str = "\n".join(html_str)
    return  html.Div([
                dcc.Graph(figure=fig),
                html.Hr(),
                html.H3("가장큰 그룹은 "+str(group)+"이며 평균은 "+str(m)+" 입니다"),
                html.H6(html_str)
            ])




# application 구동(web server)
app.run_server(port=7775)
```
