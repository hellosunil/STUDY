```python
import pandas as pd
df = pd.read_csv('서울_시간별기상정보_20130511_20160702.csv', encoding='cp949')
```

```python
df
```

```python
df = df.drop(['기온'], axis=1)
```

```python
df = df.drop(['강수량'], axis=1)
```

```python
df = df.dropna()
```

```python
df
```

```python
df.info()
```

```python
df['시각'] = pd.to_datetime(df['시간'], format='%H:%M', errors='raise')
```

```python
df
```

```python
df.info()
```

```python
df1 = df[df['시각'].dt.hour == 9]
```

```python
df1
```

```python
df2 = df1.drop(['시각'], axis=1)
```

```python
df2
```

```python
df2['날짜'] = pd.to_datetime(df2['날짜'], format='%Y-%m-%d')
```

```python
df2 = df2.reset_index()
```

```python
df2
```

```python
df2.info()
```

```python
def nomyun(num):
    if num == 4.0 or num == 5.0:
        return 1
    else:
        return 0
```

```python
df2['지면_func'] = df2['지면상태'].apply(nomyun)
```

```python
df2
```

```python
df_11 = df2[df2['날짜'].dt.month == 11]
```

```python
df_12 = df2[df2['날짜'].dt.month == 12]
```

```python
df_1 = df2[df2['날짜'].dt.month == 1]
```

```python
df_2 = df2[df2['날짜'].dt.month == 2]
```

```python
df_3 = df2[df2['날짜'].dt.month == 3]
```

```python
result1 = pd.concat([df_11, df_12, df_1, df_2, df_3])
```

```python
result1
```

```python
result=result1.sort_values(by=['날짜'], axis=0)
```

```python
result1 = result.drop(['index'], axis=1)
result2 = result1.drop(['지면상태'], axis=1)

result3 = result2.reset_index(drop=True)
result3
```

```python
result3.to_csv('0900_지면상태_11to3월_기간.txt', encoding='utf-8')
```
